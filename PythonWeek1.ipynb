## To explain // it is helpful to look at the *floor* function which can be found in the *math* library.

```python
import math
print("math.floor(12.3) =",math.floor(12.3)); print("math.floor(-12.1) =",math.floor(-12.1))
```

## As you can see *floor(x)* return the biggest integer not exceeding *x*. If *x* is positive that amounts to dropping the digits to the right if the decimal point, e.g in *12.3* we just drop *.3*. For negative numbers we drop the fractional part and then move one unit down, e.g. in *-12.1* we drop the *.1* to get *-12* then go one step down to *-13*
## *x//y* is the floor of *x/y*. i.e. it is the largest integer not exceeding *x/y* (at least in most cases). Roughly the biggest number that can be multiplied by *y* without exceeding x.
## x%y is the difference between *x* and *(x//y)\*y*

```python
# Some examples of // and % for you to try
print("10//3 =",10//3) # should be 3
print("-10//3 =",-10//3) #should be -4
print("10%3 =",10%3) #should be 1
print("-10%3 =",-10%3) # shoulld be 2 which is -10-(-4)*3
```

## Don't worry too much about memorizing what // and % do.
## However, you need to remember the order in which the operations are executed. This can be remembered as PEMDAS parenthesis ,i.e. ( ), then exponentiation \*\* , then unary minus (- with single operand) then multiplication and dividision (\*,/,//,%)  executed from left to right then addition and subtraction from left to right

## You can always use psrenthesis to specify how you want the expression to be evaluated

```python
print("10+5*3 =",10+5*3) #multiplication is performed first
print("(10+5)*3 =",(10+5)*3) # parenthesis is evaluated first
print("10**2*3 =",10**2*3) # exponentiation is performed first
print("10/2*3 =",10/2*3) # multiplication and division are done left to right
```

```python
print("2**3**2 =",2**3**2) # exponentiation left to right
print("-10**2 =",-10**2) # exponentiation is evaluated before unary minus
```

# Boolean expressions

## In programmming **conditional execution** is fundamental. Consider the problem of finding the larger of two numbers *a* and *b* and use this as the value of *max*. This is imprtant in sorting , the arrangemt of records by a number, e.g. the arrangement of student records by student id number, which is why students are given id numbers. Returning to computing *max* we set *max* to be *a* if *a>b* and to *b* otherwise. Thus the value assigned to *max* is dependent on a whether a condition *a>b* is true or not. Building conditions based on comapring values of numerical variable is very important in programming since what we want  the program to do might depend of the whether a condition is true or false. The building blocks of boolean expression are comparison operators. The code below illustartes the comaprison operators

```python
x=10;y=2;
print("x>y is ",x>y) #x larger than y
print("x>=5*y is ",x>=5*y)  # x larger than or equal y
print("x==y is ",x==y) # x equals y
print("x!=y is ",x!=y) #x is not equal y
print("x<y is " ,x<y) #
print ("x<=y is ",x<=y)
```

# String indexing

## Strings in python are a sequence of characters indexed from 0. The character at index *i* of a string *s* is expressed as *s[i]*. Try the code below

```python
st="abcd"
print(st[0]) #The first character which is "a"
print(st[2]) # The third charcater which is "c"
print("My Example"[3]) #The fourth charcater of "My Example" which is "E"
```

## We can specify several consecutive characters in a string by using *[i:j]* this specifies the charcters with index starting with *i* and ending with *j-1* <u> Note the character at index j is not included</u>. The following code illustrates this **slicing** ,i.e. taking a slice from a string.

```python
st="0123456" # so st[i]="i"
print("st[2:4] is  ",st[2:4]) # Note "4"=st[4] will not be included
print("st[:3] is ",st[:3]) # [:j] is treated as [0:j] so the slice start from the first charcter and ends at charcater j-2
print("st[2:]  is  ",st[2:]) #[i:] will result in starting the slice at i and continuing to the end of the string
```
